```python
doc = pymupdf.open("Resume.pdf")
for text in doc:
    text = text.get_text()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  # Remove URLs
    text = re.sub(r'\S+@\S+', '[EMAIL]', text)           # Replace emails
    text = re.sub(r'\+?\d[\d -]{8,}\d', '[PHONE]', text) # Replace phone numbers
    text = " ".join(text.split())                        # Remove extra whitespaces
    text = text.lower()                                  # Optional -- lower case text
    print(text)
```

## Building Section Classifier

```python
import fitz  # PyMuPDF
import os

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Example usage:
text = extract_text_from_pdf("Resume.pdf")
print(text[:500])  # preview
```

```python
from tqdm import tqdm

# Directory containing 2000 PDFs
pdf_dir = r"E:\ResumeParser.ai\notebooks\data"
output_dir = "./extracted_texts"
os.makedirs(output_dir, exist_ok=True)

for pdf_file in tqdm(os.listdir(pdf_dir)):
    if pdf_file.endswith(".pdf"):
        pdf_path = os.path.join(pdf_dir, pdf_file)
        text = extract_text_from_pdf(pdf_path)
        with open(os.path.join(output_dir, pdf_file.replace(".pdf", ".txt")), "w", encoding="utf-8") as f:
            f.write(text)
```

```python
def split_into_chunks(text):
    lines = text.splitlines()
    chunks = [line.strip() for line in lines if line.strip()]
    return chunks
```

```python
import glob

all_chunks = []

txt_files = glob.glob("./extracted_texts/*.txt")
for path in tqdm(txt_files):
    with open(path, encoding="utf-8") as f:
        text = f.read()
    chunks = split_into_chunks(text)
    all_chunks.extend(chunks)
```

```python
def label_chunk(text):
    t = text.lower()
    if any(k in t for k in ["bachelor", "master", "university", "grade", "gpa"]):
        return "Education"
    if any(k in t for k in ["worked as", "responsibilities", "experience", "internship", "role"]):
        return "Experience"
    if any(k in t for k in ["python", "sql", "c++", "skills", "excel", "tensorflow"]):
        return "Skills"
    if any(k in t for k in ["project", "developed", "implemented"]):
        return "Projects"
    if any(k in t for k in ["certificate", "certification", "course", "training"]):
        return "Courses & Certificates"
    if any(k in t for k in ["hobbies", "interests", "football", "music"]):
        return "Hobbies"
    if any(k in t for k in ["achievement", "award", "winner", "runner-up"]):
        return "Achievements"
    return "Unknown"
```

```python
labeled_chunks = []
for c in all_chunks:
    label = label_chunk(c)
    if label != "Unknown":  # skip unclassified for now
        labeled_chunks.append({"text": c, "label": label})

print(f"Total labeled chunks: {len(labeled_chunks)}")
```

```python
import pandas as pd

df = pd.DataFrame(labeled_chunks)
df.to_csv("labeled_chunks.csv", index=False)
```

```python
df = pd.read_csv("labeled_chunks.csv")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df["text"],
    df["label"],
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=5000, ngram_range=(1,2))),
    ("clf", LogisticRegression(max_iter=500))
])

pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```
