```python
from fuzzywuzzy import fuzz
def detect_sections(text, threshold=90):

    sections = {}
    current_section = None
    lines = text.splitlines()

    for raw_line in lines:
        line = raw_line.strip()
        if not line:
            continue  # skip empty lines

        normalized_line = re.sub(r"\s+", " ", line).strip().lower()

        # Check if this line is a heading
        found_section = None
        best_score = 0
        for section, keywords in SECTION_KEYWORDS.items():
            for keyword in keywords:
                score = fuzz.ratio(normalized_line, keyword.lower())
                if score > best_score:
                    best_score = score
                    found_section = section

        if best_score >= threshold:
            # New heading detected
            current_section = found_section
            sections[current_section] = []
        else:
            # No heading detected
            if current_section:
                sections[current_section].append(line)
            else:
                # This line does not belong to any section
                # You can:
                # 1. Ignore
                # 2. Assign to an "unlabeled" section:
                # sections.setdefault("unlabeled", []).append(line)
                pass

    return sections



# Example usage
resume_text = """
PROJECTS 
Jan 2025 — Feb 2025 
Mushroom Classification — Machine Learning 
Developed a machine learning model in Python to classify edible vs. 
poisonous mushrooms using features like odor and gill color. Achieved 
99% accuracy with a Decision Tree classifier, demonstrating strong 
skills in data analysis, feature engineering, and model training. 
Jan 2025 — Jan 2025 
Library Management — Python Project 
Developed a Python-based Library Management System with a user-
friendly GUI to handle adding, borrowing, and returning books. 
Integrated email database functionality to track user transactions and 
generate real-time listings of available books. 
 
COURSES & CERTIFICATES 
Jun 2025 — Jun 2025 
Microsoft Power BI — Infosys Springboard 
Jun 2025  
FinQuest — Department of Finance Studies, University of Delhi 
Jun 2025 
Central India Hackathon 2.0 — Unstop 
Aug 2023 — Jan 2025 
Full Stack Data Science Pro — PW Skills 
Oct 2024 
Codemathon 2024 — Department of Mathematics, NIT Kurukshetra 
"""

sections = detect_sections(resume_text)
print(sections)
```

```python
from fuzzywuzzy import fuzz

def detect_heading(line, threshold=85):
    """
    Checks if the given line matches any known section heading.
    Returns the normalized section key if match found, else None.
    """
    cleaned_line = line.strip().lower()
    best_match = None
    best_score = 0

    for section, keywords in SECTION_KEYWORDS.items():
        for keyword in keywords:
            keyword_clean = keyword.lower()
            score = fuzz.ratio(cleaned_line, keyword_clean)
            if score > best_score:
                best_score = score
                best_match = section

    if best_score >= threshold:
        return best_match
    else:
        return None
```

```python
import re

def group_project_lines(lines):
    projects = []
    buffer = []

    date_pattern = re.compile(r"\b(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s+\d{4}")

    for line in lines:
        clean_line = line.strip()
        if not clean_line:
            continue  # skip empty lines

        # Check if this line is a date line
        if date_pattern.search(clean_line):
            # If buffer has previous project, save it
            if buffer:
                projects.append("\n".join(buffer).strip())
                buffer = []
            buffer.append(clean_line)
        else:
            buffer.append(clean_line)

    # Append the last project
    if buffer:
        projects.append("\n".join(buffer).strip())

    return projects
```

```python
grouped_projects = group_project_lines(parsed_data)

for p in grouped_projects:
    print("==== PROJECT ====")
    print(p)
    print()
```

```python
def label_section(text):
    lower = text.lower()

    if any(keyword in lower for keyword in ["bachelor", "master", "university", "school", "grade", "percentage"]):
        return "Education"
    elif any(keyword in lower for keyword in ["worked as", "intern", "experience", "trainee", "responsibility"]):
        return "Experience"
    elif any(keyword in lower for keyword in ["python", "machine learning", "deep learning", "sql", "c++", "skills"]):
        return "Skills"
    elif any(keyword in lower for keyword in ["project", "developed", "implemented", "created"]):
        return "Projects"
    elif any(keyword in lower for keyword in ["certificate", "certification", "course"]):
        return "Courses & Certificates"
    elif any(keyword in lower for keyword in ["hobby", "interest", "football", "reading"]):
        return "Hobbies"
    elif any(keyword in lower for keyword in ["achievement", "winner", "runner-up", "hackathon"]):
        return "Achievements"
    else:
        return "Unknown"
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

labels = ["Education", "Experience", "Skills", "Projects", "Courses & Certificates", "Hobbies", "Achievements"]

text = """
Developed a machine learning model in Python to classify edible vs. poisonous mushrooms.
"""

result = classifier(text, candidate_labels=labels)
print(result)
```
